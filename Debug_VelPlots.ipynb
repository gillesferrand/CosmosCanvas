```python
# Import
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# Import our package and the colourspace package
import velmap as vmap
from colourspace import maps
```

```python
# Create the Default Colour Map. 

# Set min/max values
min_value = -1.0
max_value = 1.0

# Chroma max: 
c_max = 30.

# Create divergent velocity cmap.
# This uses default  ...
cmap = "CC-vmap-default"
vmap.create_cmap_velocity(min_value,max_value,name=cmap, Cval_max=c_max)
# By default the colourmap is registered to the mpl namespace
```

```python
maps.test_cmaps(names=[cmap],nsteps=[17],subplots=(1,1),figsize=(15,6),fname="plots/testcmap_CC-velmap")
```

```python
# Let's show the default colour map in LCH space, again rescaled to 0:1 using the plot_path function.
axes = maps.plot_path(cmap, space='LCH', stack='H', axes=[]  , styles=['-' ], legend_label=cmap)
```
